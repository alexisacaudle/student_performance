```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
students = pd.read_csv('students.csv')

# Print first few rows of data
print(students.head())

# Print summary statistics for all columns
print(students.describe(include='all'))
print('More students live in urban areas than rural.\n')
```

```python
# Calculate mean
mean_math_grade = students['math_grade'].mean()
print(f'The average grade in math is {mean_math_grade}\n')

# Calculate median
med_math_grade = students['math_grade'].median()
print(f'The median math grade is {med_math_grade}.')
if med_math_grade > mean_math_grade:
  print('The median math grade is larger than the average math grade.\n')
else:
  print('The median math grade is smaller than the average amth grade.')

# Calculate mode
mode_math_grade = students['math_grade'].mode()[0]

print(f'The most common math grade is {mode_math_grade}.')
print('The most common math grade is lower than the average and median math grades.')

# Calculate range
range_math_grade = students['math_grade'].max() - students['math_grade'].min()
print(f'The range of math grades is {range_math_grade}\n.')

# Calculate standard deviation
std_math_grade = students['math_grade'].std()
print(f'The standard deviation for math grades is {std_math_grade}.')
print('The standard deviation is about 4.6, while the average grade is about 10.4. This means that about two thirds of students are earning a grade between 5.8 (calculated as 10.4 - 4.6) and 15 (calculated as 10.4 + 4.6).\n')

# Calculate MAD
mad_math_grade = (students['math_grade'] - students['math_grade'].mean()).abs().mean()

print(f"The difference between each student's score and the average score is {mad_math_grade}.")
```

```python
# Create a histogram of math grades
sns.histplot(x='math_grade', data=students)
plt.show()
plt.clf()
```
